```python
bank_branches = {"MG Road", "Indiranagar", "Whitefield", "Electronic City"}

customers = {"Rahul", "Anita", "Kiran", "Sneha", "Vikram"}

new_customers = {"Amit", "Priya", "Rahul", "Divya"}

loan_customers = {"Rahul", "Kiran", "Asha", "Ramesh"}

savings_customers = {"Rahul", "Sneha", "Priya", "Nikhil"}

transaction_ids = {101, 102, 103, 104, 105}

balances = {85000, 120000, 45000, 95000, 275000}

cities = {"Bangalore", "Hyderabad", "Chennai", "Pune"}

new_cities = {"Mumbai", "Bangalore", "Delhi"}

account_types = {"Savings", "Current", "Loan"}
```

```python
### Basic Operations
# 1. Print the `customers` set.
print(customers)
# 2. Find the number of elements in the `customers` set.
print(len(customers))
# 3. Check whether "Rahul" exists in the `customers` set.
a={"Rahul"}
b=a.issubset(customers)
print(b)
# 4. Check whether "Meena" exists in the `customers` set.
a={"Meena"}
b=a.issubset(customers)
print(b)
# 5. Iterate through all elements in the `bank_branches` set.
for branch in bank_branches:
  print(branch)
```

```python
# Add "Meena" to the customers set.
# Add "Mysore" to the cities set.
customers.add("Meena")
cities.add("Mysore")
print(customers)
print(cities)
```

```python
# Update the customers set with new_customers.
# Update the cities set with new_cities.
print(customers)
print(new_customers)
a=customers.update(new_customers)
print(customers)
print(cities)
print(new_cities)
cities.update(new_cities)
print(cities)
```

```python
# Remove "Rahul" from the customers set.
print(customers)
customers.remove('Rahul')
print(customers)
```

```python
# Remove "Pune" from the cities set.
print(cities)
cities.remove('Pune')
print(cities)
```

```python
# Discard "Anita" from the customers set.
# Discard "Delhi" from the cities set.
# Discard "Kolkata" from the cities set.
customers.discard('Anita')
print(customers)
cities.discard('Delhi')
print(cities)
cities.discard('Kolkata')
print(cities)
```

```python
# Remove a random element from the transaction_ids set using pop().
# Remove another random element from the account_types set.
# Remove all elements from the balances set.
print(transaction_ids)
transaction_ids.pop()
print(transaction_ids)
print(account_types)
account_types.pop()
print(account_types)
print(balances)
balances.clear()
print(balances)
```

```python
# Create a copy of the customers set.
# Modify the copied set and verify that the original set remains unchanged.
customers_copy=customers.copy()
print(customers_copy)
customers_copy.add("Rohit")
print(customers_copy)
print(customers)
```

```python
# Find the union of customers and new_customers.
print(customers)
print(new_customers)
a=customers.union(new_customers)
print(a)
```

```python
# Find the common customers between customers and new_customers.
# Find customers who have both Savings and Loan accounts.
print(customers)
print(new_customers)
a=customers.intersection(new_customers)
print(a)
print(savings_customers)
print(loan_customers)
b=savings_customers.intersection(loan_customers)
print(b)
```

```python
# Find customers present only in customers but not in new_customers.
# Find customers who have only Savings accounts.
print(customers)
print(new_customers)
a=customers.difference(new_customers)
print(a)
print(savings_customers)
print(loan_customers)
b=savings_customers.difference(loan_customers)
print(b)
```

```python
# Find customers that are present in either customers or new_customers but not both.
print(customers)
print(new_customers)
a=customers.symmetric_difference(new_customers)
print(a)
```

```python
# Check whether {"Rahul", "Kiran"} is a subset of loan_customers.
print(loan_customers)
a={"Rahul","Kiran"}
b=a.issubset(loan_customers)
print(b)
```

```python
# Check whether customers is a superset of {"Rahul", "Anita"}.
print(customers)
a={"Rahul","Anita"}
b=customers.issuperset(a)
print(b)
```

```python
# Check whether loan_customers and new_cities are disjoint.
a=loan_customers.isdisjoint(new_cities)
print(a)
```

```python
# Find the maximum balance in the balances set.
print(balances)
a=max(balances)
print(a)
# Find the minimum balance in the balances set.
b=min(balances)
print(b)
# Find the total of all balances.
c=sum(balances)
print(c)
# Sort the balances set in ascending order.
d=sorted(balances)
print(d)
# Convert the customers set into a list.
print(customers)
e=list(customers)
print(e)
# Convert the cities set into a tuple.
print(cities)
f=tuple(cities)
print(f)
# Remove duplicates from the list: ["Rahul", "Rahul", "Anita", "Sneha", "Sneha"] by converting it into a set.
h=["Rahul", "Rahul", "Anita", "Sneha", "Sneha"]
print(h)
i=set(h)
print(i)
```
